```python
резюмируем вышепроделланное в компактном коде
 - читаем файл, делим данные по sep, указываем кодировку, обрабатываем данные в столбце 'Date', указываем что первым
    дате будет день, и индексами будет данные их столбца 'Date'
 - создаем из полученного ДФ новый, с данными по конкретной дорожке, добавляем к каждой дате соттветсвующий ей день недели
 - группируем данные по суммарному количеству велосипедистов на дорожке по дням недели
 - даём цифровым значениям дней недели 0-6 естественные обозначения
 - выводим данные на график
```

```python
bikes = pd.read_csv(r'c:\Users\1cons\Desktop\bikes.csv', 
                    sep=';', encoding='latin1', 
                    parse_dates=['Date'], dayfirst=True, 
                    index_col='Date')
# Add the weekday column
berri_bikes = bikes[['Berri 1']].copy()
berri_bikes.loc[:,'weekday'] = berri_bikes.index.weekday

# Add up the number of cyclists by weekday, and plot!
weekday_counts = berri_bikes.groupby('weekday').aggregate(sum)
weekday_counts.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_counts.plot(kind='bar')
```

Шаблон для получения данных о погоде в Канаде, конкретно по ID в Монтреале. Берем за март 2012.

```python
url_template = "http://climate.weather.gc.ca/climate_data/bulk_data_e.html?format=csv&stationID=5415&Year={year}&Month={month}&timeframe=1&submit=Download+Data"
url = url_template.format(month=3, year=2012)
weather_mar2012 = pd.read_csv(url, index_col='Date/Time', parse_dates=True, encoding='latin1')
```

```python
weather_mar2012
```

Удаляем столбцы (axis=1), в которых отсутствуют данные (dropna) с хотя бы одним полем с NaN (how='any') 

```python
weather_mar2012 = weather_mar2012.dropna(axis=1, how='any')
weather_mar2012[:5]
```

Удаляем (drop) столбцы с неинтересующими и несущественными данными

```python
weather_mar2012 = weather_mar2012.drop(['ï»¿"Longitude (x)"', 'Latitude (y)', 'Station Name', 'Climate ID', 'Year', 'Month', 'Day', 'Time'], axis=1)
```

Облагораживаем имена столбцов

```python
weather_mar2012.columns = [u'Temp (C)', 
    u'Dew Point Temp (C)', u'Rel Hum (%)', u'Wind Spd (km/h)', u'Visibility (km)', u'Stn Press (kPa)', u'Weather']
```

```python
weather_mar2012[:5]
```

Создаем копированием новый ДФ с данными по температуре
Добавляем столбец "Часы"
Группируем по часам данные по средней температуре и строи график

```python
temperatures = weather_mar2012[[u'Temp (C)']].copy()
temperatures.loc[:,'Hour'] = weather_mar2012.index.hour
temperatures.groupby('Hour').aggregate(np.median).plot()
```

Обобщая вышепроделанные операции, определяем функцию для считывания данных с аргументами (год, месяц)

```python
def download_weather_month(year, month):
    url_template = "http://climate.weather.gc.ca/climate_data/bulk_data_e.html?format=csv&stationID=5415&Year={year}&Month={month}&timeframe=1&submit=Download+Data"
    url = url_template.format(year=year, month=month)
    weather_data = pd.read_csv(url, index_col='Date/Time', parse_dates=True, encoding='latin1')
    weather_data = weather_data.dropna(axis=1)
    weather_data = weather_data.drop(['ï»¿"Longitude (x)"', 'Latitude (y)', 'Station Name', 'Climate ID', 'Year', 'Month', 'Day', 'Time'], axis=1)
    weather_data.columns = [u'Temp (C)', 
    u'Dew Point Temp (C)', u'Rel Hum (%)', u'Wind Spd (km/h)', u'Visibility (km)', u'Stn Press (kPa)', u'Weather']
    return weather_data
```

```python
download_weather_month(2012, 12)[:5]
```

Объединяем данные всех месяцев за год

```python
data_by_month = [download_weather_month(2012, i) for i in range(1, 13)]
```

Объединение данных в ДФ за год (pd.concat)

```python
weather_2012 = pd.concat(data_by_month)
weather_2012
```

сохраним результат в CSV файл

```python
weather_2012.to_csv(r'c:\Users\1cons\Desktop\weather_2012.csv')
```

```python
# Строим графики температур и снежности вместе
temperature = weather_2012['Temp (C)'].resample('M').median()
is_snowing = weather_2012['Weather'].str.contains('Snow')
snowiness = is_snowing.astype(int).resample('M').mean()

# Name the columns
temperature.name = "Temperature"
snowiness.name = "Snowiness"
```

```python
stats = pd.concat([temperature, snowiness], axis=1)
stats
```

Из-за разных масштабов данных нагляднее вывести данные на двух отдельных графиках

```python
stats.plot(kind='bar', subplots=True, figsize=(15, 10))
```
