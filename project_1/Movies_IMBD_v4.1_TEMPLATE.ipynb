Вопрос 14. Какой режиссер снял больше всего фильмов в стиле Action?

```python
data14 = pd.read_csv('movie_bd_v5.csv')
data14['genres'] = data14['genres'].str.split('|')
data14['director'] = data14['director'].str.split('|')
data14 = data14.explode('genres')
data14 = data14.explode('director')
display(data14)
data14_res = data14[data14['genres'] == 'Action'].groupby(['director'])['genres'].count().sort_values(ascending=False)
display(data14_res)
```

Вопрос 15. Фильмы с каким актером принесли самые высокие кассовые сборы в 2012 году?

```python
data15 = pd.read_csv('movie_bd_v5.csv')
data15['cast'] = data15['cast'].str.split('|')
data15_cast = data15.explode('cast')
display(data15_cast)
data15_res = data15_cast[data15_cast['release_year'] == 2012].groupby(['cast'])['revenue'].sum().sort_values(ascending=False)
display(data15_res)
```

Вопрос 16. Какой актер снялся в большем количестве высокобюджетных фильмов?

```python
data16 = data15_cast.copy()
data16_res = data16[data16['budget'] > data16['budget'].mean()].groupby(['cast'])['cast'].count().sort_values(ascending=False)
display(data16_res)
```

Вопрос 23. Для какого режиссера зима — самое продуктивное время года?
